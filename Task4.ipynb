```python
# Import necessary libraries
import sklearn.datasets as datasets
import pandas as pd 
import matplotlib.pyplot as plt
```

## Importing and examining the Dataset

```python
# Import dataset
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()

y = iris.target
print(y)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Training the algorithm

```python
from sklearn.model_selection import train_test_split
df_train, df_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=0)
```

## Defining the Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(df, y)
```

```python
y_pred = dtree.predict(df_test)
```

```python
# Creating the confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

## Visualizing the Decision Tree

```python
from sklearn import tree
```

```python
fn = ['sepal length(cm)', 'sepal width(cm)', 'petal length(cm)', 'petal width(cm)']
cn = ['setosa', 'versicolor', 'virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)

tree.plot_tree(dtree, feature_names = fn, class_names = cn, filled = True);
```
